```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset=pd.read_csv('../Data/Salary_dataset_DS.csv.csv')

dataset.info()
df=dataset.copy()

```

```python


df=df.drop(columns='Unnamed: 0')
df.info()
```

```python
#checking for the misssing values
print("The missing values in each columns")
for i in df.columns:
    print(f"{i}-{df[i].isnull().sum()}")
```

```python
#identifing the unique values in each for categorial separation

for i in df.columns:
    print(f"The unique values {i} : {df[i].unique()} ")

    print()
    print()
```

```python
df.info()
df.describe()
df.shape
df.head()
```

```python
#removing unneceessary columns
df_updated=df.drop(columns=['salary','salary_currency']).copy()
df_updated.info()
```

```python
#droping the NAN Vaues

df_updated.dropna(inplace=True)
df_updated.info()
```

```python
#checking for the duplicated values
duplicate_sum=df_updated.duplicated().sum()
print(duplicate_sum)


df_updated.drop_duplicates(inplace=True)
```

```python
# duplicate_sum=df_updated.duplicated().sum()
# print(duplicate_sum)
```

```python
#checking the dataset that it is positive or negative 


# df_updated=df_updated[df_updated['salary_in_usd']>0]

df_updated=df_updated[(df_updated['remote_ratio']>=0) & (df_updated['remote_ratio'])<=100 ]
df_updated.info()
```

```python
#renaming the columns for the vaild dataset
df_updated=df_updated.rename(columns={
    "work_year":"Work_year_representation",
    "job_title ":"Name_Of_Job",
    "salary_in_usd":"Salary"
})

df_updated.info()
```

```python
#resetting the index for the proper dataset

df_updated.reset_index(drop=True)
```

```python
#detecting outlier using the the box plot approach

sns.boxplot(x=df_updated['salary_in_usd'])
plt.xlabel("Values in the dataset")
plt.xlim(0,800000)
plt.show()
```

```python
#final  Checkup

print(df_updated.shape)
print(df_updated.isnull().sum())
print(df_updated.duplicated().sum())


"Final the dataset is clean and can be saved as an csv file"
```

```python
df_updated.to_csv("../Data/Cleaned_salary_DS.csv",index=False)
```
