```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder

#getting the clean dataset

df=pd.read_csv("../Data/Cleaned_salary_DS.csv")

df.info()
```

```python
#separating the columns in the dataset

y_pred=df['Salary']
x_pred=df.drop(columns="Salary",axis=1)

```

```python
x_pred.info()
```

```python
#identifying the categorical values
categorical_column=[]
for i in x_pred.columns:
    if x_pred[i].dtype=="object":
        categorical_column.append(i)

print(categorical_column)
```

```python
#Applying the label encoder to convert the text data to numeric data

label_encode={}
for i in categorical_column:

    encode=LabelEncoder()

    x_pred[i]=encode.fit_transform(x_pred[i])

    label_encode[i]=encode  

label_encode 

```

```python
#creating new folder  of "model" to save the Encoder
# import os
# import pickle

# os.makedirs("model", exist_ok=True)

# with open("model/label_encoders.pkl", "wb") as f:
#     pickle.dump(label_encode, f)

x_pred.to_csv("../Data/X_processed.csv", index=False)
y_pred.to_csv("../Data/y_processed.csv", index=False)

```
